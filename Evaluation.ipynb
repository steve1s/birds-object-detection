# Run evaluation script

The evaluation script from the Tensorflow Object Detection API will be run via the command listed below. Results for object detection measures such Mean Average Precision (mAP), mAP at Intersection over Union (IOU) 0.5 and mAP at IOU, 0.75, will be provided. These metrics can show how well your model is doing when tested against test data. We must specify the label map path in the configuration and input data into the eval_input_reader field. You must enter the path to the model's configuration file, model directory, and checkpoint directory in the command below.

```python
!CUDA_VISIBLE_DEVICES="" python model_main_tf2.py --pipeline_config_path=/home/msc1/Desktop/Tensorflow-Object-Detection-API/Base/v1/object_detection/training/TF2/training/{PATH_TO_MODEL}/pipeline.config --model_dir=/home/msc1/Desktop/Tensorflow-Object-Detection-API/Base/v1/object_detection/training/TF2/training/{PATH_TO_MODEL} --checkpoint_dir=/home/msc1/Desktop/Tensorflow-Object-Detection-API/Base/v1/object_detection/training/TF2/training/{PATH_TO_MODEL} --alsologtostderr
```
