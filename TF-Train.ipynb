# Hyperparameters and Configuration of the config file

For hyperparameters, the Tensorflow Object Detection API uses a configuration file. You can change every hyperparameter for your object detection task in this file. The command below opens a code editor with the configuration file. The fields below will be changed for this project.

* The number of classes you want the model to be trained on is indicated by the parameter num_classes. Since we are training the model on three classes, it will be set to 3.
* batch_size: The quantity of samples the network trains on at once. It can be boosted by having additional memory capacity. Because of the high resolution of our photos, Faster R-CNN requires greater processing power. I'll therefore leave this number at 1.
* num_steps: The number of steps the model will be trained using. I'll begin the training with 12000 steps and use Tensorboard to examine the loss experienced during training to determine whether or not the model needs to be trained further.
* fine_tune_checkpoint: The location of the pretrained model's checkpoint file.
* path to the label_map.pbtxt file, label_map_path.
* input_path: The train TFRecord file's location.
